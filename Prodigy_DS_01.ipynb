```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("cleaned_diabetes_dataset.csv")

sns.set_style("whitegrid")

plt.figure(figsize=(10, 6))
sns.histplot(df['patient_age'], bins=10, kde=True, color='blue')
plt.title('Distribution of Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

#read your data file
df = pd.read_csv('cleaned_diabetes_dataset.csv')

# Count the frequency of each diabetes level
level_counts = df['level_of_diabetes'].value_counts()

# Create the pie chart
colors= ['red', 'green', 'blue']
plt.pie(level_counts, labels=level_counts.index, colors= colors, autopct='%1.1f%%', startangle=90,) #autopct='%1.1f%%' formats the percentage labels to one decimal place.
plt.title('Distribution of Diabetes Levels')
plt.axis('equal')  # Equal aspect ratio ensures a circular pie chart
plt.show()
```
